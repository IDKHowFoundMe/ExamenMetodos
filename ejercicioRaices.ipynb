```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return x**5 + 5.25 *x **4 + 4.125 * x**3 - 9.125 * x**2 -14.625 * x - 5.625
x = np.linspace(-4,2,200)

plt.plot(x, f(x), color="red", label="f(x)")
plt.axhline(0, color='black',linewidth=0.5, ls='-')
plt.xlabel("x")
plt.ylabel("y")

plt.show()
```

```python
import scipy.optimize as opt


def fprime(x):
    return 5 * x**4 + 21.0 * x**3 + 12.375 * x**2 - 18.25 * x - 14.625
    


root = opt.bisect(f, -1,-0.5)

print(f"Raíz encontrada por bisección en el intervalo [-1.5, -0.5]: {root}")
for i in range(-2,-1):
    if f(i)*f(i+1)<0:
        root = opt.bisect(f, i, i+1)
        
        print(f"Raíz encontrada por bisección en el intervalo [{i}, {i+1}]: {root}")


for i in range(-20,20):
    if f(i)*f(i+1)<0:
        root = opt.newton(f, i, fprime=fprime)
        print(root)    
        
```
