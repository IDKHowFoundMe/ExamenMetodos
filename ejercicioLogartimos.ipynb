```python
import math 
def raizRedondeda5(r: float)->float: 
    for i in range(5):
        r=round(math.sqrt(r),6)
    return r


print(raizRedondeda5(2))
```

```python
for i in range(2,15):
    logbrigs= 32 * ((raizRedondeda5(i)-1)/math.log(10))
    errorRelativo = abs((logbrigs - math.log10(i)/ math.log(10)))  
    print (i)
    print(f"logbrigs {logbrigs} errorRelativo: {errorRelativo}")
    
```
